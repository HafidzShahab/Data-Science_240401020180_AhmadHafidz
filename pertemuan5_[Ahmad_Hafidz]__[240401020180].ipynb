```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', palette='Set2')
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
print(df.shape)
print(df.dtypes)
print(df.describe())
```

```python
avg_bill = df.groupby('day')['total_bill'].mean()

plt.figure(figsize=(8,5))
avg_bill.plot(kind='bar', color='skyblue')

plt.title('Rata-rata Total Tagihan per Hari')
plt.xlabel('Hari')
plt.ylabel('Total Bill (USD)')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.histplot(
    data=df,
    x='total_bill',
    kde=True,
    color='teal'
)

plt.title('Distribusi Total Bill')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(
    data=df,
    x='sex',
    y='total_bill',
    palette='Set2'
)

plt.title('Distribusi Total Bill Berdasarkan Gender')
plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    size='size',
    sizes=(30,200),
    alpha=0.7
)

plt.title('Hubungan Total Bill dan Tip')
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(14,10))

# Grafik 1
avg_bill.plot(kind='bar', ax=axes[0,0], color='skyblue')
axes[0,0].set_title('Avg Bill per Day')

# Grafik 2
sns.histplot(data=df, x='total_bill', kde=True, ax=axes[0,1])

# Grafik 3
sns.boxplot(data=df, x='sex', y='total_bill', ax=axes[1,0])

# Grafik 4
sns.scatterplot(
    data=df,
    x='total_bill',
    y='tip',
    hue='time',
    ax=axes[1,1]
)

plt.tight_layout()
plt.show()
```
