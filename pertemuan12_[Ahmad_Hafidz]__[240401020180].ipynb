```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules

from sklearn.metrics.pairwise import cosine_similarity
```

```python
transaksi = [
    ['Roti','Susu','Telur'],
    ['Roti','Selai'],
    ['Roti','Susu','Selai'],
    ['Susu','Keju'],
    ['Roti','Mentega'],
    ['Kopi','Gula'],
    ['Teh','Gula'],
    ['Roti','Selai','Mentega'],
    ['Sereal','Susu'],
    ['Roti','Susu','Selai'],
    ['Keju','Susu'],
    ['Roti','Telur'],
    ['Kopi','Gula','Susu'],
    ['Roti','Mentega','Selai'],
    ['Sereal','Susu','Roti']
]

print(transaksi)
```

```python
te = TransactionEncoder()

te_array = te.fit(transaksi).transform(transaksi)

df = pd.DataFrame(te_array, columns=te.columns_)

df.head()
```

```python
frekuensi = df.sum().sort_values(ascending=False)

frekuensi.plot(kind='bar')

plt.title("Frekuensi Produk")

plt.ylabel("Jumlah")

plt.show()
```

```python
freq_items = apriori(
    df,
    min_support=0.20,
    use_colnames=True
)

freq_items.sort_values(
    by='support',
    ascending=False
)
```

```python
rules = association_rules(
    freq_items,
    metric='confidence',
    min_threshold=0.5
)

rules = rules[rules['lift']>1]

rules[['antecedents',
       'consequents',
       'support',
       'confidence',
       'lift']].sort_values(
           by='lift',
           ascending=False)
```

```python
katalog = pd.DataFrame({

'produk':[
'Roti',
'Selai',
'Susu',
'Sereal',
'Telur',
'Keju',
'Kopi',
'Gula',
'Teh',
'Mentega'
],

'kategori':[
'Bakery',
'Bakery',
'Dairy',
'Bakery',
'Dairy',
'Dairy',
'Minuman',
'Bumbu',
'Minuman',
'Dairy'
]

})

katalog
```

```python
fitur = pd.get_dummies(katalog['kategori'])

similarity = cosine_similarity(fitur)

similarity
```

```python
def rekomendasi(nama_produk, top_n=3):

    idx = katalog[katalog['produk']==nama_produk].index[0]

    skor = list(enumerate(similarity[idx]))

    skor = sorted(
        skor,
        key=lambda x:x[1],
        reverse=True
    )

    skor = [x for x in skor if x[0]!=idx]

    hasil = [katalog.iloc[i]['produk']
             for i,_ in skor[:top_n]]

    return hasil
```

```python
print(rekomendasi("Roti"))

print(rekomendasi("Susu"))

print(rekomendasi("Kopi"))
```

```python
produk = "Roti"

rules_roti = rules[
    rules['antecedents'].apply(
        lambda x: produk in x
    )
]

print("Association Rules")

print(rules_roti[['consequents','lift']])

print()

print("Content Based")

print(rekomendasi(produk))
```

Mahasiswa diharapkan dapat memahami bahwa:

Association Rule Mining menemukan pola barang yang sering dibeli secara bersamaan menggunakan metrik Support, Confidence, dan Lift.
Content-Based Filtering memberikan rekomendasi berdasarkan kemiripan atribut produk.
Kedua metode memiliki kelebihan dan kekurangan sehingga dalam implementasi nyata sering digabungkan menjadi Hybrid Recommendation System.
